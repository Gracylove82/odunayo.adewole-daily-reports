Exercise 1:

Create a simple line plot for y = x for x values ranging from 0 to 10.

# Exercise 1



````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(0, 11)

y = x

plt.figure(figsize = (8, 6))

plt.plot(x, y, color = "red")

plt.xlabel("x")

plt.ylabel("y")

plt.title("Plot of $y=x$")

plt.show()
````

Exercise 2:

Plot y = x^2 using Matplotlib for x values between -10 and 10.

# Exercise 2



````python
x = np.arange(-10, 11)

y = x**2

plt.figure(figsize = (8, 6))

plt.plot(x, y, color = "blue")

plt.xlabel("x")

plt.ylabel("$x^2$")

plt.title("Plot of $y=x^2$")

plt.grid(True)

plt.show()
````

Exercise 3:

Create a scatter plot with 50 random points where x and y values are between 0 and 10.

# Exercise 3



````python
np.random.seed(0)

x = np.random.randint(0, 10, size = 50)
y = np.random.randint(0, 10, size =  50)
````

````python
plt.figure(figsize = (8, 6), dpi = 100)

plt.scatter(x, y, linestyles = "-")

plt.xlabel("$X$")
plt.ylabel("$Y$")

plt.title("Scatter Plot of 50 Random Point")

plt.grid(True)

plt.show()
````

Exercise 4:

Plot a sine wave using Matplotlib for x values between 0 and 2π.

# Exercise 4



````python
x = np.linspace(0, 2*np.pi, 100)

y = np.sin(x)
plt.figure(figsize = (9, 7))
plt.plot(x, y, color = "brown")

plt.xlabel("$x$")

plt.ylabel("$y$")

plt.title("Sine Wave Plot")

plt.grid(True)

plt.show()
````

Exercise 5:

Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9].

# Exercise 5



````python
categories = ['A', 'B', 'C', 'D']

values = [3, 7, 5, 9]

plt.figure(figsize = (6, 7))

plt.bar(categories, values, color = "blue", linewidth = 2)

plt.xlabel("Categories")

plt.ylabel("Values")

plt.title("Bar Plot for Categories Versus Values")



plt.show()
````

Exercise 6:

Create a histogram of 1000 random numbers generated from a normal distribution.

# Exercise 6

````python
c = np.random.randn(1000)

plt.hist(c, bins = 30, color = "brown")

plt.xlabel("Values")

plt.ylabel("Frequency")

plt.title("Histogram of Random Normal Generated Values")

plt.show()
````

Exercise 7:

Plot two line plots on the same graph: y1 = x^2 and y2 = x^3 for x ranging from 0 to 10.

# Exercise 7

````python
x = np.linspace(0, 10)

y1 = x**2

y2 = x**3


plt.figure(figsize = (9, 6))

plt.plot(x, y1, color = "brown")

plt.plot(x, y2, color = "blue")

plt.title("Line Plot of $y1 = x^2$ and $y2 = x^3$")

plt.grid(True)

plt.show()
````

Exercise 8:

Add titles, axis labels, and a legend to the plot in Exercise 7.

# Exercise 8

````python
plt.figure(figsize = (9, 6))

plt.plot(x, y1, color = "brown", label = "y1 = x^2")

plt.plot(x, y2, color = "blue", label = "y2 = x^3")

plt.title("Line Plot of $y1 = x^2$ and $y2 = x^3$")


plt.legend()


plt.grid(True)

plt.show()
````

Exercise 9:

Create a subplot with 2 rows and 1 column. Plot y1 = x^2 in the first subplot and y2 = x^3 in the second.

# Exercise 9

````python


fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# First subplot
ax1.plot(x, y1, color="brown", label="y1 = x^2")
ax1.set_title('Plot of y1 = x^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.grid(True)


#Second subplot
ax2.plot(x, y2, color="blue", label="y2 = x^3")
ax2.set_title('Plot of y1 = x^2')
ax2.set_xlabel('x')
ax2.set_ylabel('y1')
ax2.grid(True)


````

Exercise 10:

Use a different color and linestyle for each line in a plot of three functions: y1 = x, y2 = x^2, and y3 = x^3.

# Exercise 10



````python
plt.figure(figsize = (8, 6), dpi = 100)

y1 = x

y2 = x**2

y3 = x**3


plt.plot(x, y1, color = "blue", linestyle = "-", label = "y1 = x")

plt.plot(x, y2, color = "black", linestyle = ":", label = "y2 = x^2")

plt.plot(x, y3, color = "brown", linestyle = "--", label = "y3 = x^3")

plt.grid(True)

plt.legend()

plt.show()
````
